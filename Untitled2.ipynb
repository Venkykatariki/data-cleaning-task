Step 1: Import file

```python
import pandas as pd
df = pd.read_csv("Customer Personality Analysis.csv")
```

```python
df = pd.read_csv("Customer Personality Analysis.csv", delimiter='\t')
```

```python
df.head()
df.info()
```

Step 2: Checking Missing Values.

```python
df.isnull().sum()
```

```python
# Drop rows with missing income
df = df.dropna(subset=['Income'])

# Verify again
df.isnull().sum()
```

Remove duplicate rows

```python
# Remove duplicate rows if any
df = df.drop_duplicates()

# Check shape after removing duplicates
df.shape
```

Step 3: Standardize 'Education' and 'Marital_Status'

```python
df['Education'] = df['Education'].str.strip().str.title()
df['Marital_Status'] = df['Marital_Status'].str.strip().str.title()

# Preview the cleaned text columns
df[['Education', 'Marital_Status']].drop_duplicates()
```

Step 4: Convert date formats to a consistent type

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

print(df['Dt_Customer'].head())
print(df.dtypes['Dt_Customer'])
```

Step 5: Rename column headers to be clean and uniform (e.g., lowercase, no spaces).

```python
df.columns = df.columns.str.strip() \
                       .str.lower() \
                       .str.replace(' ', '_') \
                       .str.replace('[^a-z0-9_]', '', regex=True)

print(df.columns)
```

Step 6: Check and fix data types

```python
print(df.dtypes)
```
